```python
import pandas as pd
import numpy as np
```

```python
import json
with open('dac_match_details.txt', 'r') as f:
    context = f.read()
data = json.loads(context)
df = pd.DataFrame(data)
print(len(df))
```

```python
df['radiant_name'].value_counts()
```

```python
radiant_group = df.groupby('radiant_name')#以天辉的对名分组
vg_frame = radiant_group.get_group('Vici Gaming')#选择天辉为Vici Gaming的比赛
#print(vg_frame['radiant_win'])
win_frame = vg_frame[vg_frame['radiant_win']]#VG在天辉胜利的比赛
dire_group = df.groupby('dire_name')#以夜魇的对名分组
vg_frame = dire_group.get_group('Vici Gaming')
win_frame = win_frame.append(vg_frame[vg_frame['radiant_win']==False])
print(win_frame['match_id'])
win_frame=win_frame.sort('start_time')#重新排序
print(win_frame['match_id'])
```

```python
'''
使用filter函数，根据player的account_id查找队员
以fy为例，fy的account_id为101695162
'''
def get_player(x):
    for players in x['players']:
        for player in players:
            if player['account_id'] == 101695162:
                return True
    return False
grouped = df.groupby(level=0)
playerFrame = grouped.filter(get_player)
print(playerFrame['match_id'])
```

```python
teamName = "Vici Gaming"
def isTeamWin(x):
    if x['radiant_name'].iloc[0] == teamName and x['radiant_win'].iloc[0]:
        return True
    elif x['dire_name'].iloc[0] == teamName and not x['radiant_win'].iloc[0]:
        return True
    return False
def getByTeamName(x):
    if x['radiant_name'].iloc[0] == teamName:
        return True
    elif x['dire_name'].iloc[0] == teamName:
        return True
    return False
grouped = df.groupby(level=0)
teamFrame = grouped.filter(isTeamWin)
print(teamFrame['match_id'])
```

```python
'''
Boolean indexing
使用map
'''
account_id = 101695162
teamName = "Vici Gaming"
def getByAccount_id(players):
    for player in players:
        if player['account_id'] == account_id:
            return True
    return False
playerFrame = df[df['players'].map(getByAccount_id)]
print(playerFrame['match_id'])
##根据Boolean indexing选择DataFrame
teamFrame = df[(df['radiant_name']==teamName) | (df['dire_name']==teamName)]
print(teamFrame['match_id'])
teamWinFrame = df[((df['radiant_name']==teamName)&(df['radiant_win'])) | ((df['dire_name']==teamName)&(~df['radiant_win']))]
print(teamWinFrame['match_id'])
```
