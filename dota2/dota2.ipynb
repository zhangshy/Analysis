```python
gold_per_mins = []
for players in frame['players']:
    for player in players:
        if player['account_id'] == 101695162:
            gold_per_mins.append(player['gold_per_min'])
gold_series = Series(gold_per_mins)       
gold_series.plot()
print("%d %d %d %d" % (len(gold_series), gold_series.min(), 
                       gold_series.max(), gold_series.mean()))
```

```python
cframe = frame[frame['game_mode']==2]#提取'game_mode'=2的项
print(len(cframe))
cframe['game_mode'].value_counts()
```

```python
cframe['players'][0][0]['ability_upgrades'][0]['time']#测试一下第一个比赛的第一个玩家学习1技能的时间
```

```python
def get_ability_upgrades_times(frame):
    upgrades_times = []
    for players in frame['players']:
        for player in players:
            upgrades_times.append(player['ability_upgrades'][0]['time'])
    time_series = Series(upgrades_times)
    return time_series
upgrade_time_series = get_ability_upgrades_times(cframe)
print(len(upgrade_time_series))
```
